```python
import pandas as pd
```

```python
df1 = pd.read_csv('Cardiotocographic.csv') 
```

```python
print("Missing Values Count:")
print(df1.isnull().sum())
```

```python
col1 = df1.select_dtypes(include=['int64', 'float64' , 'object']).columns
```

```python
for col in col1:
    df1[col] = df1[col].fillna(df1[col].mode()[0])
```

```python
print("Missing Values Count:")
print(df1.isnull().sum())
```

```python
print("Before correction:")
print(df1.dtypes)
```

```python
import numpy as np
def remove_outliers_iqr(dataframe, columns):
    for col in columns:
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        
        # Replace outliers with boundary values (capping)
        dataframe[col] = np.where(dataframe[col] < lower_limit, lower_limit, dataframe[col])
        dataframe[col] = np.where(dataframe[col] > upper_limit, upper_limit, dataframe[col])
    
    return dataframe

df1 = remove_outliers_iqr(df1, col1)
print("Data after removing outliers:")
print(df1.describe())
```

```python
df1.to_csv("Cardiotocographic_cleaned.csv", index=False)

print("Cleaning Completed Successfully!")
print("Cleaned file saved as: Cardiotocographic_cleaned.csv")
```
