```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cardiotocographic.csv")
```

```python
# Detect numeric and categorical columns
ncols = df.select_dtypes(include=['number']).columns
ccols = df.select_dtypes(exclude=['number']).columns
if "NSP" in df.columns:
    ccols = list(ccols) + ["NSP"]
```

```python
for col in ncols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
for col in ncols:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in ccols:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in ccols:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f"Pie Chart of {col}")
    plt.ylabel("")
    plt.show()
```

```python
#SCATTER PLOTS (PAIRWISE - VARAIBLES RELATIONSHIPS)
sns.pairplot(df[ncols])
plt.show()
```

```python
#VIOLIN PLOTS FOR NUMERICAL FEATURES
for col in ncols:
    plt.figure(figsize=(6, 4))
    sns.violinplot(y=df[col])
    plt.title(f"Violin Plot of {col}")
    plt.ylabel(col)
    plt.show()
```
